```python
import json
import re

JD_FILE = "C:/Users/Rupali/OneDrive/Desktop/RecruitVerse1/RecruitVerse1/data/job_descriptions/backend_developer_jd.txt"


def load_skills_dictionary():

    with open(
        "C:/Users/Rupali/OneDrive/Desktop/RecruitVerse1/RecruitVerse1/data/skills_dictionary/skills_dictionary.json",
        "r",
        encoding="utf-8"
    ) as file:

        return json.load(file)


def extract_skills(text):

    skills_data = load_skills_dictionary()

    found_skills = []

    text_lower = text.lower()

    for category in skills_data:

        for skill in skills_data[category]:

            if skill.lower() in text_lower:

                found_skills.append(skill)

    return list(set(found_skills))


def extract_education(text):

    education_keywords = [

        "B.Tech",
        "B.E",
        "B.Sc",
        "BCA",
        "MCA",
        "MBA",
        "M.Tech",
        "M.Sc",
        "PhD",
        "Bachelor",
        "Master"
    ]

    for edu in education_keywords:

        if edu.lower() in text.lower():

            return edu

    return None


def extract_experience(text):

    pattern = r'(\d+)\+?\s+years'

    matches = re.findall(
        pattern,
        text,
        re.IGNORECASE
    )

    if matches:

        return max(
            [int(x) for x in matches]
        )

    return 0

def save_json(data, output_file):

    with open(
        output_file,
        "w",
        encoding="utf-8"
    ) as file:

        json.dump(
            data,
            file,
            indent=4
        )

def parse_jd(file_path):

    with open(
        file_path,
        "r",
        encoding="utf-8"
    ) as file:

        text = file.read()

    job = {

        "skills":
        extract_skills(text),

        "education":
        extract_education(text),

        "experience_years":
        extract_experience(text)
    }

    return job


if __name__ == "__main__":

    result = parse_jd(JD_FILE)

    print(
        json.dumps(
            result,
            indent=4
        )
    )

    save_json(
        result,
        "C:/Users/Rupali/OneDrive/Desktop/RecruitVerse1/RecruitVerse1/data/parsed_job_descriptions/backend_developer_jd.json"
    )

    print("JD JSON Saved")
```
