```python
import re
import json
import os

def load_skills_dictionary():

    with open(
        "C:/Users/Rupali/OneDrive/Desktop/RecruitVerse1/RecruitVerse1/data/skills_dictionary/skills_dictionary.json",
        "r",
        encoding="utf-8"
    ) as file:

        return json.load(file)
    
# Resume text file
INPUT_FILE = "C:/Users/Rupali/OneDrive/Desktop/RecruitVerse1/RecruitVerse1/data/parsed_resumes/raw_text"


def extract_email(text):

    pattern = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}'

    match = re.search(pattern, text)

    if match:
        return match.group()

    return None


def extract_phone(text):

    pattern = r'(\+?\d[\d\-\s]{8,15}\d)'

    match = re.search(pattern, text)

    if match:
        return match.group()

    return None


def extract_name(text):

    lines = text.split("\n")

    for line in lines[:10]:

        line = line.strip()

        if len(line) > 3 and len(line) < 50:

            if "@" not in line:

                return line

    return "Unknown"


def extract_skills(text):

    skills_data = load_skills_dictionary()

    found_skills = []

    text_lower = text.lower()

    for category in skills_data:

        for skill in skills_data[category]:

            if skill.lower() in text_lower:

                found_skills.append(skill)

    return list(set(found_skills))

def extract_education(text):

    education_keywords = [

        "B.Tech",
        "B.E",
        "B.Sc",
        "BCA",
        "MCA",
        "MBA",
        "M.Tech",
        "M.Sc",
        "PhD",
        "Bachelor",
        "Master"
    ]

    for edu in education_keywords:

        if edu.lower() in text.lower():

            return edu

    return None

import re

def extract_experience(text):

    pattern = r'(\d+)\+?\s+years'

    matches = re.findall(
        pattern,
        text,
        re.IGNORECASE
    )

    if matches:

        return max(
            [int(x) for x in matches]
        )

    return 0

def save_json(data, output_file):

    with open(
        output_file,
        "w",
        encoding="utf-8"
    ) as file:

        json.dump(
            data,
            file,
            indent=4
        )

def parse_resume(file_path):

    with open(
        file_path,
        "r",
        encoding="utf-8"
    ) as file:

        text = file.read()

    candidate = {

    "name":
    extract_name(text),

    "email":
    extract_email(text),

    "phone":
    extract_phone(text),

    "skills":
    extract_skills(text),

    "education":
    extract_education(text),

    "experience_years":
    extract_experience(text)
}

    return candidate


def process_all_resumes():

    input_folder = r"C:\Users\Rupali\OneDrive\Desktop\RecruitVerse1\RecruitVerse1\data\parsed_resumes\raw_text"

    output_folder = r"C:\Users\Rupali\OneDrive\Desktop\RecruitVerse1\RecruitVerse1\data\parsed_resumes\parsed_json"

    os.makedirs(
        output_folder,
        exist_ok=True
    )

    count = 0

    for file in os.listdir(input_folder):

        if file.endswith(".txt"):

            file_path = os.path.join(
                input_folder,
                file
            )

            result = parse_resume(
                file_path
            )

            output_file = os.path.join(
                output_folder,
                file.replace(
                    ".txt",
                    ".json"
                )
            )

            save_json(
                result,
                output_file
            )

            count += 1

            print(
                f"Processed: {file}"
            )

    print(
        f"\nTotal JSON Files Created: {count}"
    )


if __name__ == "__main__":

    process_all_resumes()
```
