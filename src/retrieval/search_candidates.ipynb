```python
import psycopg2

conn = psycopg2.connect(

    database="recruitverse",

    user="postgres",

    password="root123",

    host="localhost",

    port="5432"

)

cursor = conn.cursor()
```

```python
def search_by_skill(skill):

    cursor.execute(

        """

        SELECT DISTINCT

            c.candidate_id,
            c.name,
            c.experience_years

        FROM candidates c

        JOIN candidate_skills s

        ON c.candidate_id = s.candidate_id

        WHERE LOWER(s.skill)=LOWER(%s)

        LIMIT 20

        """,

        (skill,)

    )

    return cursor.fetchall()
```

```python
results = search_by_skill("Python")

for row in results:

    print(row)
```

```python
def search_by_experience(years):

    cursor.execute(

        """

        SELECT

            candidate_id,
            name,
            experience_years

        FROM candidates

        WHERE experience_years >= %s

        LIMIT 20

        """,

        (years,)

    )

    return cursor.fetchall()
```

```python
results = search_by_experience(5)

for row in results:

    print(row)
```

```python
def get_candidate(candidate_id):

    cursor.execute(

        """

        SELECT *

        FROM candidates

        WHERE candidate_id=%s

        """,

        (candidate_id,)

    )

    return cursor.fetchone()
```

```python
print(

    get_candidate(1)

)
```

```python
def get_candidate_skills(candidate_id):

    cursor.execute(

        """

        SELECT skill

        FROM candidate_skills

        WHERE candidate_id=%s

        """,

        (candidate_id,)

    )

    return cursor.fetchall()
```

```python
while True:

    print("\n")

    print("1. Search Skill")

    print("2. Search Experience")

    print("3. Candidate Details")

    print("4. Exit")

    choice = input("Choose: ")

    if choice == "1":

        skill = input(
            "Skill: "
        )

        results = search_by_skill(
            skill
        )

        for row in results:

            print(row)

    elif choice == "2":

        years = int(
            input(
                "Experience: "
            )
        )

        results = search_by_experience(
            years
        )

        for row in results:

            print(row)

    elif choice == "3":

        cid = int(
            input(
                "Candidate ID: "
            )
        )

        print(
            get_candidate(cid)
        )

        print(
            get_candidate_skills(cid)
        )

    elif choice == "4":

        break
```
