```python
import json
import os
import psycopg2

conn = psycopg2.connect(

    database="recruitverse",

    user="postgres",

    password="root123",

    host="localhost",

    port="5432"
)

cursor = conn.cursor()
```

```python
def insert_job(job, job_title):

    cursor.execute(

        """

        INSERT INTO jobs(

            job_title,
            education,
            experience_years

        )

        VALUES(

            %s,
            %s,
            %s

        )

        RETURNING job_id

        """,

        (

            job_title,

            job.get("education"),

            job.get("experience_years")

        )

    )

    job_id = cursor.fetchone()[0]

    insert_job_skills(

        job_id,

        job.get("skills", [])

    )
```

```python
def insert_job_skills(job_id, skills):

    for skill in skills:

        cursor.execute(

            """

            INSERT INTO job_skills(

                job_id,
                skill

            )

            VALUES(

                %s,
                %s

            )

            """,

            (

                job_id,

                skill

            )

        )
```

```python
def process_all_jobs():

    input_folder = r"C:\Users\Rupali\OneDrive\Desktop\RecruitVerse1\RecruitVerse1\data\parsed_job_descriptions"

    count = 0

    for file in os.listdir(input_folder):

        if file.endswith(".json"):

            file_path = os.path.join(
                input_folder,
                file
            )

            with open(
                file_path,
                "r",
                encoding="utf-8"
            ) as f:

                job = json.load(f)

            job_title = file.replace(
                ".json",
                ""
            )

            insert_job(
                job,
                job_title
            )

            count += 1

            print(
                f"Inserted: {file}"
            )

    conn.commit()

    print(
        f"\nTotal Jobs Inserted: {count}"
    )
```

```python
if __name__ == "__main__":

    process_all_jobs()

    conn.close()
```
